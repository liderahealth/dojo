### import

```
import pandas as pd
from pysus.utilities.readdbc import read_dbc
```

### SIHSUS

```
RDES1801 = 'RDES1801.dbc'
RDES1802 = 'RDES1802.dbc'
RDES1803 = 'RDES1803.dbc'
RDES1804 = 'RDES1804.dbc'
RDES1806 = 'RDES1806.dbc'
RDES1807 = 'RDES1807.dbc'
RDES1808 = 'RDES1808.dbc'
RDES1809 = 'RDES1809.dbc'
RDES1810 = 'RDES1810.dbc'
RDES1811 = 'RDES1811.dbc'
RDES1812 = 'RDES1812.dbc'

SIHES1801 = read_dbc(RDES1801, encoding='iso-8859-1')
SIHES1802 = read_dbc(RDES1802, encoding='iso-8859-1')
SIHES1803 = read_dbc(RDES1803, encoding='iso-8859-1')
SIHES1804 = read_dbc(RDES1804, encoding='iso-8859-1')
SIHES1805 = read_dbc(RDES1805, encoding='iso-8859-1')
SIHES1806 = read_dbc(RDES1806, encoding='iso-8859-1')
SIHES1807 = read_dbc(RDES1807, encoding='iso-8859-1')
SIHES1808 = read_dbc(RDES1808, encoding='iso-8859-1')
SIHES1809 = read_dbc(RDES1809, encoding='iso-8859-1')
SIHES1810 = read_dbc(RDES1810, encoding='iso-8859-1')
SIHES1811 = read_dbc(RDES1811, encoding='iso-8859-1')
SIHES1812 = read_dbc(RDES1812, encoding='iso-8859-1')

SIHES = pd.concat([ SIHES1801,
                  SIHES1802,
                  SIHES1803,
                  SIHES1804,
                  SIHES1805,
                  SIHES1806,
                  SIHES1807,
                  SIHES1808,
                  SIHES1809,
                  SIHES1810,
                  SIHES1811,
                  SIHES1812 ])
```
