## Tipos de Acidente de trabalho por ano (INSS) no Brasil

```python
df_act07 = pd.read_csv('ACT07.csv', encoding = 'latin-1')
df_act07.head()
```

```python
df_act07['CID_split'] = df_act07['CID'].apply(lambda s: s.split(':'))
```

```python
df_act07.head()
```

```python
df_cid_id_name = df_act07['CID_split'].apply(pd.Series)
```

```python
df_cid_id_name = pd.merge(df_act07, pd.DataFrame(df_cid_id_name), left_index=True, right_index=True)
```

```python
#df_cid_id_name["_cid_id"]=df_cid_id_name["_cid_id"]=""
#df_cid_id_name["_cid_name"]=""
df_cid_id_name.rename(columns={0:'_cid_id',1:'_cid_name'},inplace=True)
```

```python
df_cid_id_name.head()
```

```python
df_cid_id_name['_cid_id'].value_counts()
```

```python
df_dicid = pd.read_csv('http://bit.ly/dicid')
```

```python
df_dicid.head()
```

```python
df_cid_id_name['_cid_id'] = df_cid_id_name['_cid_id'].apply(lambda s: s if len(s) == 4 else s if s in ['998', '999'] else s+'0')
```

```python
df_dicid_id_name = pd.merge(df_cid_id_name, df_dicid[['subcat', 'description']], how='inner', left_on='_cid_id', right_on='subcat', )
df_dicid_id_name.shape
```

## Quantos acidentes de trabalho por ano no brasil?
